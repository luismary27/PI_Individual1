Se elimina registros nulos

```python
steam_games_df1 = steam_games_df1.dropna(how='all', subset=['id']) 
```

```python
#verificamos que se elimino correctamente
val_null=steam_games_df1.isnull().sum()
print(val_null)
```

```python
#vemos los tipos de datos que esta cada columna
steam_games_df1.dtypes
```

```python
steam_games_df1.shape
```

```python
#cambiamos el id en tipo entero ya que esta en flotate
steam_games_df1['id'] = steam_games_df1['id'].astype('Int64')
```

Algunas columnas estan anidadas

```python
#Desanidamos la columna genres
steam_games_df1['genres'] = steam_games_df1['genres'].apply(eval)
lista_genres = list(set(genre for sublist in steam_games_df1['genres'] for genre in sublist))
```

```python
lis=lista_genres
df_genres= pd.DataFrame(lis)
df_genres
```

```python
#Aplicamos el metodo .explode para trabajar con cada valor de la lista de forma individual
steam_games_df2 = steam_games_df1.explode("genres")
```

```python
#Revisamos que se realizo correctamente
steam_games_df2
```

```python
steam_games_df2['release_date'] =pd.to_datetime(steam_games_df2['release_date'], errors= 'coerce')
steam_games_df2['release_date'] =steam_games_df2['release_date'].dt.year.fillna(0).astype(int)
```

```python
steam_games_df2.dtypes
```

```python
#Reemplazamos los valores de price
steam_games_df2.replace({"free to use": 0, "free to play": 0}, inplace=True)
```

```python
steam_games_df3 = steam_games_df2.dropna(subset=['id'])
steam_games_df3.shape
```

```python
steam_games_df3['id']=steam_games_df3['id'].apply(np.int64)
steam_games_df3= steam_games_df3.rename(columns={"id":"item_id"})
```
